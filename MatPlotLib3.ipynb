```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
profit=[10,20,30,40]
labels=["Amazon","Flipkart","Snapdeal","JioMart"]

colors=["red","cyan","yellow","blue"] #create list of colors then pass as argument in the .pie function .

explode=[0.5,0.2,0,0]

plt.pie(profit,colors=colors,labels=labels,autopct="%.2f%%",explode=explode,counterclock=False,startangle=90) #pie chart plotting with colors,labels,percentage,explode,generally pie chart follows counter clock direction to make it false we pass it and the startangle is shift by 90 degree.
plt.show()
```
