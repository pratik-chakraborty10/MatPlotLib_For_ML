```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10,0.25)
y=x**4
#plt.scatter(x,y)
plt.plot(x,y,color="green",marker="*",linewidth=1)
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
y2=x**3

plt.plot(x,y,color="red",label="x^4")
plt.plot(x,y2,color="blue",label="x^3")
plt.legend() #to display the labels for better understanding the  different color scheme data.
plt.xlabel("X",fontsize=20) # x axis label
plt.ylabel("Y",fontsize=20) # y axis label
plt.title("My chart") #title of the graph
plt.axis([0,20,0,10000]) #x axis going from 0 to 20 and y axis going from o to 10000
plt.grid() #enables the grid line
plt.text(10,9000,"Pratik") #to write something random in the graph at a particular axis point.
plt.show()
```
